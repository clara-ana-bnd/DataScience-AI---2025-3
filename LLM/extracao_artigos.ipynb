```python
import pandas as pd
import os
```

```python
pasta = r'Metais'

dataframes = []

for arquivo in os.listdir(pasta):
    caminho_arquivo = os.path.join(pasta, arquivo)
    df = pd.read_excel(caminho_arquivo)[:100]
    df['Arquivo_origem'] = arquivo
    dataframes.append(df)
```

```python
df_final = pd.concat(dataframes, ignore_index=True)
df_final.info()
df_final.to_excel("dataframe_artigos_reduzido.xlsx")
```

```python
# Sua lista (por exemplo, uma lista de listas)
lista_de_dados = [
    ['Nome', 'Idade', 'Cidade'],
    ['Alice', 30, 'SÃ£o Paulo'],
    ['Bob', 25, 'Rio de Janeiro'],
    ['Charlie', 35, 'Belo Horizonte']
]
import csv
# Abre o arquivo em modo de escrita ('w') e usa 'with' para garantir que ele seja fechado automaticamente
with open('nome_arquivo', 'w', newline='', encoding='utf-8') as arquivo_csv:
    # Cria um objeto de escrita
    escritor_csv = csv.writer(arquivo_csv)
    
    # Escreve cada linha da lista no arquivo CSV
    for linha in lista_de_dados:
        escritor_csv.writerow(linha)

print(f"Lista salva com sucesso em '{'nome_arquivo'}'")
```
